pd.read_csv() 与 pd.read_table() 的区别:
官方文档解释：

pd.read_csv():Read a comma-separated values (csv) file into DataFrame.
将逗号分隔值（csv）文件读入 DataFrame。

pd.read_table():Read general delimited file into DataFrame.
将通用分隔文件读入 DataFrame。

二者间的互相转换：
pd.read_csv() -> pd.read_table(): pd.read_csv(filepath, sep = '\t')
pd.read_table() -> pd.read_csv(): pd.read_table(filepath, sep = ',')

tsv:用制表符（Tab,'\t'）作为字段值的分隔符, pd.read_table()读入
csv:是用半角逗号（','）作为字段值的分隔符, pd.read_csv()读入

reset_index: Reset the index of the DataFrame, and use the default one instead. If the DataFrame has a MultiIndex, this method can remove one or more levels.
