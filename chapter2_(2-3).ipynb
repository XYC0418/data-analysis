## 第二节 数据重构1

```python
#读入数据
df_1 = pd.read_csv('data/train-left-up.csv')
df_1.head()
```

```python
df_2 = pd.read_csv('data/train-left-down.csv')
df_2.head()
```

```python
df_3 = pd.read_csv('data/train-right-up.csv')
df_3.head()
```

```python
df_4 = pd.read_csv('data/train-right-down.csv')
df_4.head()
```

```python
#使用concat 将 train-left-up.csv 和 trian-right-up.csv 横向合并并保存为result_up
result_up = pd.concat([df_1, df_3], axis=1)
result_up.head()
```

```python
result_down = pd.concat([df_2, df_3], axis=1)
result_down.head()
```

```python
result = pd.concat([result_up, result_down], axis=0)
result.head()
```

pandas.concat(objs, axis=0, join='outer', ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=False, copy=True)

https://pandas.pydata.org/docs/reference/api/pandas.concat.html?highlight=concat#

objs(a sequence or mapping of Series or DataFrame objects): 需要合并的对象。

join('inner' or 'outer'): 如何处理轴上的索引

axis({0/'index', 1/'columns'}, default 0): 0为纵向合并，1为横向合并。

ignore_index: 为True时，不使用原来的索引而将合并的标为(0,1,...,n-1)。

keys(sequence, default None): 在外层构建层次索引。

names(list, default None): 生成层次索引的名称。

```python
result_up = df_1.join(df_3)
result_up.head()
```

```python
result_down = df_2.join(df_4)
result_down.head()
```

DataFrame.join(other, on=None, how='left', lsuffix='', rsuffix='', sort=False)

other(DataFrame, Series, or list of DataFrame): 需要和DataFrame 合并的表。

how({‘left’, ‘right’, ‘outer’, ‘inner'}, default 'left'): 合并方式。

```python
result = result_up.append(result_down)
result.head()
```

```python
result_up = pd.merge(df_1, df_3, left_index=True, right_index=True)
result_up.head()
```

```python
result_down = pd.merge(df_2, df_4, left_index=True, right_index=True)
result_down.head()
```

```python
result = result_up.append(result_down)
result.head()
```

merge: 应用场景一般为两张表有相同的列，类似于SQL 中的JOIN函数。没有相同的列时需要讲将left_index 和 right_index 设为True。

join: 拼接列。如果两个表有重复的列名，需指定lsuffix, rsuffix参数。

concat:按轴进行合并。

```python
result.to_csv("result.csv")
```

```python
unit_result = result.stack()
unit_result.head()
```

DataFrame.stack(level=- 1, dropna=True) 

将列旋转为行。

返回与当前 DataFrame 相比具有一个或多个新的最内层的多级索引的重构 DataFrame 或 Series。通过旋转当前数据框的列来创建新的最内层：

如果列有一个级别，则输出是一个Series；

如果列有多个级别，则新的索引级别是从规定的级别获取的，并且输出是一个 DataFrame。

```python
#计算男性和女性的平均票价
fare_sex = result.groupby('Sex')['Fare'].mean()
fare_sex
```

```python
#统计男女的存活人数
survived_sex = result.groupby(['Sex'])['Survived'].sum()
survived_sex
```

```python
#计算客舱不同等级的存活人数
survived_Pclass = result['Survived'].groupby(result['Pclass']).sum()
survived_Pclass
```

女性的存活率高于男性，客舱等级高的存活率越高。

```python
#运用agg方法实现
result.groupby('Sex').agg({'Fare': 'mean', 'Survived': 'sum'}).rename(columns=
                            {'Fare': 'mean_fare', 'Survived': 'sum_survived'})
```

```python
result.groupby('Pclass').agg({'Survived': 'sum'}).rename(columns={'Survived': 'sum_survived'})
```

```python
#统计不同等级的票中的不同年龄的船票花费的平均值。
result.groupby(['Pclass', 'Age'])['Fare'].mean()
```
